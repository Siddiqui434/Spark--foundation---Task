### Profits in States

```python
# state counts
print('Total number of States are', data['State'].nunique())
print(data['State'].value_counts())
```

```python
# sorting the profits of states in decreasing order
states_profit = data.groupby(by ='State').sum()['Profit'].sort_values(ascending = False).reset_index()
```

```python
# plotting top 10 states with highest profits
plt.figure(figsize = (15,10))
plt.title('States with maximum profit', size = 30)
sns.barplot(x = 'State', y = 'Profit', data = states_profit.head(10))
plt.xticks(rotation = 60, size = 15)
plt.yticks(size = 15)
plt.xlabel('States', size = 20)
plt.ylabel('Profits', size = 20)
plt.show()
```

```python
# plotting bottom 10 states with least profits
plt.figure(figsize = (15,10))
plt.title('States with minimum profits', size = 30)
sns.barplot(x = 'State', y = 'Profit',data = states_profit.tail(10))
plt.xticks(rotation = 60, size = 15)
plt.yticks(size = 15)
plt.xlabel('States', size = 20)
plt.ylabel('Profits', size = 20)
plt.show()
```



The most profitable state is Califonia and the least profitable state is Texas



## profits in cities

```python
# sorting the profits of cities in decreasing order
cities_profit = data.groupby(by ='City').sum()['Profit'].sort_values(ascending = False).reset_index()
```

```python
# plotting top 10 city with highest profits
plt.figure(figsize = (15,10))
plt.title('Cities with maximum profits', size = 30)
sns.barplot(x = 'City', y = 'Profit',data = cities_profit.head(10))
plt.xticks(rotation = 60, size = 15)
plt.yticks(size = 15)
plt.xlabel('Cities', size = 20)
plt.ylabel('Profits', size = 20)
plt.show()
```

```python
# plotting top 10 cities with least profits
plt.figure(figsize = (15,10))
plt.title('Cities with minimum profit', size = 30)
sns.barplot(x = 'City', y = 'Profit',data = cities_profit.tail(10))
plt.xticks(rotation = 60, size = 15)
plt.yticks(size = 15)
plt.xlabel('cities', size = 20)
plt.ylabel('Profits', size = 20)
plt.show()
```

The most profitable city is New York City and the least profitable city is Philadelphia.

## Category Plots

```python
sns.countplot(x = 'Category',data = data)
plt.title('Category counts',size = 20);
```

office supplies are in high demand.Technology category has the least supply.

```python
category_profit = data.groupby(by ='Category').sum()['Profit'].sort_values(ascending = False)
plt.figure(figsize = (6,6))
plt.title('Profit in terms of Categories', size = 20)
category_profit.plot(kind = 'bar')
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.xlabel('Categories')
plt.ylabel('Profits')
plt.show()
```

Technology categories is the most profitable, despite having the least supply. Furniture categories gains much less profit as compared to other two sectors.

```python
# top 10 states with highest sales
sales = data.groupby('Category').sum()['Sales'].nlargest(n=10)
# top 10 states with highest profit
profit = data.groupby('Category').sum()['Profit'].nlargest(n=10)

plt.style .use ('seaborn')
sales.plot(kind = 'bar',fontsize = 15, color = 'c')
profit.plot(kind = 'bar',fontsize = 15, color = 'm')
plt.xlabel('Categories',fontsize = 15)
plt.ylabel('Total Sales and Profits', fontsize = 15)
plt.title("Sales vs Profits in terms of Categories", size = 20)
plt.show()
```

Technology and office Supplies have high sales and profit.

## sub - category plots

```python
plt.figure(figsize = (16,8))
sns.countplot(x = 'Sub-Category', data = data)
plt.title('Sub-Category counts', size = 20);
```

Binders and Paper have very demand. Machines and Copiers have the least demand.

```python
sub_profit = data.groupby('Sub-Category').sum()['Profit'].sort_values(ascending = False)

plt.figure(figsize = (16,6))
plt.title('Profit in terms of Sub-Categories', size = 20)
sub_profit.plot(kind = 'bar')
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.xlabel('Sub-Categories')
plt.ylabel('Profits')
plt.show()
```

Copiers have highest profit, while Tables have lowest profit.

```python
min_profit = data.groupby('Sub-Category').min()['Profit'].sort_values()

plt.figure(figsize = (8,6))
min_profit.plot()
plt.title('Sub-Categories with Highest Losses', size = 20)
plt.xticks(rotation = 60, fontsize = 15)
plt.yticks(fontsize = 15)
plt.show()

```

Machines have incurred more losses than other sub- categories.

## Shipping Mode Plots

```python
sns.countplot(x='Ship Mode', data = data)
plt.title('Shipping Mode Counts', size = 20);
```

Standard Class is the most preferred shipping mode.

```python
ship_profit = data.groupby('Ship Mode').sum()['Profit'].sort_values(ascending = False)


plt.figure(figsize = (16,6))
plt.title('Profit in terms of Shipping Modes', size = 20)
ship_profit.plot(kind = 'bar')
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.xlabel('Ship-mode')
plt.ylabel('Profits')
plt.show()
```

Standard Class shipping mode shows very high profit as compared to others.

```python
ship_profit_avg = data.groupby('Ship Mode').mean()['Profit'].sort_values(ascending = False)


plt.figure(figsize = (16,6))
plt.title(' Average Profits in terms of Shipping Modes', size = 20)
ship_profit_avg.plot(kind = 'bar')
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.xlabel('avg Ship-mode')
plt.ylabel('Profits')
plt.show()
```

```python
ship_discount = data.groupby('Ship Mode').sum()['Discount'].sort_values(ascending = False).reset_index()


plt.figure(figsize = (10,6))
plt.title('Discount in terms of Shipping Modes', size = 20)
sns.barplot(x = 'Ship Mode', y = 'Discount', data = ship_discount )
plt.xticks(fontsize = 15)
plt.yticks(fontsize = 15)
plt.xlabel('Ship-mode')
plt.ylabel('discount')
plt.show()
```

Standard Class shipping mode provides maximum discount and gains maximum profit. but the average profit is minimum. 

## Segment plots

```python
sns.countplot(x = 'Segment',data = data)
plt.title('Segment counts',size = 20);
```

Consumer orders are the most and home office orders are the least.

```python
sales = data.groupby('Segment').mean()['Sales'].sort_values(ascending = False)

profits = data.groupby('Segment').mean()['Profit'].sort_values(ascending = False)



plt.style .use ('seaborn')
sales.plot(kind = 'bar',fontsize = 15, color = 'r')
profits.plot(kind = 'bar',fontsize = 15, color = 'g')
plt.xlabel('Segments',fontsize = 15)
plt.ylabel('Total Sales and Profits', fontsize = 15)
plt.title("Sales vs Profits in terms of Segments", size = 20)
plt.show()
```

Home office segment has more sales and profits, inspite of having less orders. consumer segment has the least sales and profit.

```python
plt.style.use('seaborn')
data.plot(kind = 'scatter',x = 'Sales',y = 'Profit',c ='Discount',s = 20,fontsize = 16, colormap = 'Spectral')
plt.ylabel('Total profits',fontsize = 15)
plt.title('sales vs profits on Discounts', fontsize = 20)
plt.show()
```

```python
plt.figure(figsize= (7,6))
sns.kdeplot(data['Sales'],color = 'b',label = 'Sales', shade = True, bw_method=20)
sns.kdeplot(data['Profit'],color = 'r',label = 'Profit', shade = True, bw_method=20)
plt.title('KDE Plot of Sales and Profit', size = 20)
plt.xlabel('Sales/Profit',size =  15)
plt.ylabel('Density',size = 15)
plt.legend()
plt.show()
```

Profit is more than Sales for a given value.
